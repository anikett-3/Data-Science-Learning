## 📦 11. Reshaping
| Method      | Purpose            |
| ----------- | ------------------ |
| `reshape()` | Change array shape |

> RULE--
- Rows × Columns = Total Elements

- **Reshaping is technique to change the shape or dimension of an any array without modifying the data**  
- eg --> 
`suppose we gonna conert 1D ARRAY into -> 2D ARRAY and vice versa, But we don't want to modify the data .`

- #--> `reshaping means changing the dimension of the dataset of withthout  modifly original data`.
- #--> `if there is 4 element in data after reshaping the data also 4 elements in data.`
- #--> `it return the view  of dataset.` 
### --> **it does not copy the data  --> it returns a view**

- **NOTE** --> `we only reshape only when dimension is match. `

- SYNTAX--> 
- #--> **`reshape(rows, column) `** specify new shape

- #--> **`arr.reshape(row,column)`**

````python
# EXAMPLE-->>
import numpy as np
arr= np.array([1,2,3,4,5,6])  # here dimension is match => 3+3 = 6 element
# if we pick (5,7,9),etc element and reshape into (2,3) then it show ❌ ERROr

reshaped_arr = arr.reshape(2,3)   # 2 row , 3 column
print(reshaped_arr)
````

## 📦 15. Broadcasting
## Rule 1 --> Matching Shapes

If both arrays have the same shape, NumPy performs **normal element-wise operations**.

Example

```python
[1,2,3] + [4,5,6]

↓

[5,7,9]
```

---

## Rule 2 --> Expanding Single Element

If one value (or dimension) is `1`, NumPy automatically expands it.

Example

```python
[1,2,3] + [10]

↓

[11,12,13]
```

---

## Rule 3 --> Incompatible Shapes

If shapes are not compatible, NumPy raises an error.

Example

```python
(2,3)

+

(2,)
```

❌ ValueError

````python
# Same shape/Matching shape -
(3,) + (3,)

# Compatible shape --
(2,3)+(3,)

# Incompatible shape --
(2,3)+(2,)
````
