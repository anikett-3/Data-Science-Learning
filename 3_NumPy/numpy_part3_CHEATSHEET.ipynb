## 📦 15. Broadcasting
## Rule 1 --> Matching Shapes

If both arrays have the same shape, NumPy performs **normal element-wise operations**.

Example

```python
[1,2,3] + [4,5,6]

↓

[5,7,9]
```

---

## Rule 2 --> Expanding Single Element

If one value (or dimension) is `1`, NumPy automatically expands it.

Example

```python
[1,2,3] + [10]

↓

[11,12,13]
```

---

## Rule 3 --> Incompatible Shapes

If shapes are not compatible, NumPy raises an error.

Example

```python
(2,3)

+

(2,)
```

❌ ValueError

````python
# Same shape/Matching shape -
(3,) + (3,)

# Compatible shape --
(2,3)+(3,)

# Incompatible shape --
(2,3)+(2,)
````

## 📦 18. View vs Copy
| Operation   | Returns          |
| ----------- | ---------------- |
| `reshape()` | View *(usually)* |
| `ravel()`   | View             |
| `flatten()` | Copy             |
| `copy()`    | Copy             |
| `=`         | Same Object      |
| Slicing     | View             |


## 📦 19. Attribute vs Method
| Attribute | Method       |
| --------- | ------------ |
| `.shape`  | `.reshape()` |
| `.size`   | `.astype()`  |
| `.dtype`  | `.copy()`    |
| `.ndim`   | `.flatten()` |
| `.T`      | `.ravel()`   |


### Remember

**`Attributes → No ()`**

**`Methods → Use ()`**
