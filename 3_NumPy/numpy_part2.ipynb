# 2. Replacing missing value with NUMBERs.

> Beacuse sometime in Mathematical operation wherever is nan value is then this operation is not complete.

1. Also models of machine learning can not considiring nan value.
2. ( when we call nan_to_num() it put some value and defalut value is 0)

3. syntax--->> 
>> np.nan_to_num(array, nan = value)       | default = 0

```python

arr1 = np.array([1,2,3,np.nan,5,np.nan,8])

# print(np.isnan(arr1))

cleaned_arr1 = np.nan_to_num(arr1, nan = 123)
print(cleaned_arr1)

# output - > [  1.   2.   3. 123.   5. 123.   8.]
```

## QUE 1. How to calculate important numbers based on data sets, how to use various Python modules and how to make functions that are able to predict the outcome based on what we have learned (Small Dataset).

#### SOLUTON ----
We'll use a small sales dataset that includes information about products, quantities sold, and prices.


## MODULE USED--->>
1. Pandas (used to create and manipulate the dataset.)

2. NumPy (utilized to compute statistical metrics like mean, median, and standard deviation).


`-->  Here we  demonstrates how NumPy and Pandas can be combined to perform data analysis on real-world datasets.`

```python

import pandas as pd
import numpy as np


# Sample sales data
data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Smartwatch'],
    'Units_Sold': [100, 250, 150, 300, 200],
    'Unit_Price': [1000, 500, 300, 100, 200]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate total sales for each product
df['Total_Sales'] = df['Units_Sold'] * df['Unit_Price']


# Using NumPy for calculations
mean_sales = np.mean(df['Total_Sales'])
median_sales = np.median(df['Total_Sales'])
std_dev_sales = np.std(df['Total_Sales'])

print("Mean Sales:", mean_sales)
print("Median Sales:", median_sales)
print("Standard Deviation of Sales:", std_dev_sales)


print("\nFinal Sales Data:") 
print(df)
```

## QUE 2. How can we get Big Data Sets?

##### SOLUTION

* Modules Used:
1. NumPy – to generate large/random/normal datasets.
2. Pandas – to manage and explore structured data.
3. Matplotlib – to visualize distributions and scatter plots.

`This practical shows how to work with large amounts of generated data and visualize its patterns.`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate a "big" random dataset (1000 samples)
data_size = 1000
ages = np.random.randint(18, 70, size=data_size)
incomes = np.random.normal(loc=50000, scale=15000, size=data_size)


# Create DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Income': incomes
})

print(df)
```

```python
# 1. Data Distribution: Histogram of Age
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 2. Normal Distribution (Income is generated as normal distribution)
plt.figure(figsize=(6,4))
plt.hist(df['Income'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Income Distribution (Normal)')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 3. Scatter Plot between Age and Income
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Income'], alpha=0.5, color='purple')
plt.title('Scatter Plot: Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```
