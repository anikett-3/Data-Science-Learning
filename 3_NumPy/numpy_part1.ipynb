# 🚀 NumPy - Introduction

## What is NumPy?

**NumPy (Numerical Python)** is one of the most popular Python libraries used for **fast numerical computing**. It is specially designed to perform calculations on **large datasets** efficiently using **NumPy Arrays**.

Instead of relying on Python lists and loops, NumPy performs operations using optimized C code behind the scenes, making calculations significantly faster and more memory-efficient.

---

## History of NumPy

NumPy was created by **Travis Oliphant** in **2005**.

Before NumPy, Python programmers mainly used **lists** and **loops** for mathematical calculations. Although Python is easy to use, this approach became very slow when working with millions of data points.

To solve this performance problem, Travis Oliphant developed NumPy, introducing the powerful **ndarray (N-dimensional Array)** data structure.

---

## Why was NumPy created?

Imagine you are calculating the average temperature of:

- 6 cities ✅
- 10,000 cities
- 10 lakh cities
- 5 crore sensor readings

Using Python lists and loops becomes increasingly slow as the amount of data grows.

NumPy solves this problem by storing data efficiently in arrays and performing operations using optimized compiled code.

---

## When should we use NumPy?

### Use Python Lists when

- Dataset is small.
- General-purpose programming.
- Different data types are required.

### Use NumPy when

- Dataset is large.
- Mathematical calculations are needed.
- Machine Learning
- Data Science
- Artificial Intelligence
- Scientific Computing
- Image Processing
- Financial Analysis
- Medical Research

# Python List vs NumPy Array

## Python List

A **Python list** is a built-in data structure used to store multiple values in a single variable.

Lists are flexible because they can store different data types, but they are **not optimized for large-scale numerical computations**.

---

## NumPy Array

A **NumPy Array (`ndarray`)** is the fundamental data structure provided by the NumPy library.

Unlike Python lists, NumPy arrays are designed specifically for **fast numerical computation**.

They store elements efficiently in memory and allow mathematical operations to be performed on entire arrays without writing explicit loops.

---

## Python List vs NumPy Array

| Python List | NumPy Array |
|-------------|-------------|
| Built into Python | Provided by NumPy |
| Can store different data types | Usually stores one data type (homogeneous) |
| Slower for numerical computation | Optimized for fast numerical computation |
| Uses more memory | More memory efficient |
| Requires loops for many mathematical operations | Supports vectorized operations |

---

## Why does the output look different?

Python List

```python
[1, 2, 3, 4, 5]
```

NumPy Array

```python
[1 2 3 4 5]
```

The missing commas **do not improve speed**. They are simply part of NumPy's display format.

The real performance improvement comes from:

- Efficient memory layout (contiguous memory)
- Homogeneous data type
- Optimized C implementation
- Vectorized operations

The display format and execution speed are unrelated.

# 11. COPYING ARRAYS (Assignment, View & Deep Copy)

## NOTE -->

When we create another array from an existing array, there are **3 ways**:

1. **Assignment (`=`)** → Same memory (Reference)
2. **Slicing (`arr[:]`)** → View (Shallow Copy)
3. **`copy()`** → Deep Copy (New Memory)

---

````python
## Example 1 --> Assignment (`=`)
import numpy as np

arr = np.arange(1, 21)

b = arr      # Assignment

b[0] = 9999

print("Original Array :", arr)
print("Assigned Array :", b)
````

````python
b = arr.copy()   # Deep copy (independent)
c = arr.view()   # Shallow copy (linked)
````

### Explanation -->

* `b = arr` does **not** create a new array.
* Both variables point to the **same memory**.
* Any change in `b` also changes `arr`.

````python
## Example 2 --> Slicing (View / Shallow Copy)

arr = np.arange(1, 21)

slice_arr = arr[:5]

slice_arr[0] = 100  # it also effect original array

print("Original Array :", arr)
print("Slice Array    :", slice_arr)
print("Original Array :", arr)
````

### Explanation -->

* Slicing creates a **View (Shallow Copy)**.
* A view shares the same memory as the original array.
* Changing the slice also changes the original array.


````python
# Example 3 Deep Copy --> (copy()) 
arr = np.arange(1, 21)

b = arr.copy()

b[0] = 9999

print("Original Array :", arr)
print("Copied Array   :", b)
````



### Explanation -->

* `copy()` creates a completely new array.
* Original and copied arrays have different memory locations.
* Changing one array does **not** affect the other.
---


## NOTE -->

| Method             | Memory      | Original Array Changes? |
| ------------------ | ----------- | ----------------------- |
| `b = arr`          | Same        | ✅ Yes                   |
| `arr[:]` / Slicing | Same (View) | ✅ Yes                   |
| `arr.copy()`       | New         | ❌ No                    |

# 14. Flattening Array



## NOTE -->

Now we learn how to convert a **Multi-dimensional array → 1D array**.

There are two methods:

   - Changes may affect the original array.

2. `flatten()`
   - Returns a **Copy** of the original array.
   - Changes do **not** affect the original array.

````python
import numpy as np

arr_2d = np.array([
    [1, 3, 4, 2],
    [5, 6, 7, 8]
])

print(arr_2d.ravel())

print(arr_2d.flatten())
````

````python
a = np.array([[1,2],[3,4]])

b = a.ravel()

b[0] = 100

print(a)
print(b)


# Now a also changes because b is a view.
````

## Explanation -->

### `ravel()`

**Syntax**

```python
arr.ravel()
```

- Converts a multi-dimensional array into a **1D array**.
- Returns a **View** whenever possible.
- Original array **may be affected** if the returned array is modified.

---

### `flatten()`

**Syntax**

```python
arr.flatten()
```

- Converts a multi-dimensional array into a **1D array**.
- Always returns a **Copy**.
- Original array is **not affected**.

| `flatten()` | Copy | Not affected |

# 📘 21 . Broadcasting 

## NOTE -->

Broadcasting is a NumPy feature that allows -  arrays of **different different but compatible shapes** to perform operations **without using loops**.

Advantages:

- No loops required.
- Less code.
- Faster execution.
- Automatic element-wise operations whenever possible.

- > Braodcasting is a features of  Numpy  where we perform differrent operation on different different shapes of an arrray witout using loops
- altough size of array is small or big Numpy do it quickly and easily.
- **here we don't write more line of code**

- `QUESTION` HOW NUMPY HANDLE DIFFERENT SHAPES OF ARRAY ?

- > Basically there are 3 rule for broadcasting 

1. Matching dimension == means if both array is same shape then we apply element wise operartion normally.
eg-->> [1,2,3] + [4,5,6] == [5,7,9]


2. Expanding single element --> if one array  has in one dimension then numpy expand it so it  match from another array.
eg-- a = [1,2,3]
.    b = [10 ]
so here numpy can add (2nd array single elemnet) 10 on every element of 1st array.  
then -->> output is , c= [11,12,13] 



3. Incompatitable shape   -->> if shape of two array doesnot match then it show error.
eg -- a= [1,2,3]
.     b = [4,5]  numpy cant perfomr any operatiion because it shape does not match.
  '''


````python
# Using Python Loop

prices = [100, 200, 300]

discount = 10

final_prices = []

for price in prices:
    final_p = price - (price * discount / 100)
    final_prices.append(final_p)

print(final_prices)
````

## NOTE -->

Using loops works correctly, but it becomes slow for large datasets.

````python
# Using NumPy Broadcasting

import numpy as np

prices = np.array([100, 200, 300])

discount = 10

final_prices = prices - (prices * discount / 100)

print(final_prices)
````

## Explanation -->

Here `discount` is a **scalar value**.

NumPy automatically applies the operation to every element of the array.

No loop is required.

# Broadcasting Rules

## Rule 1 --> Matching Shapes

If both arrays have the same shape, NumPy performs **normal element-wise operations**.

Example

```python
[1,2,3] + [4,5,6]

↓

[5,7,9]
```

---

## Rule 2 --> Expanding Single Element

If one value (or dimension) is `1`, NumPy automatically expands it.

Example

```python
[1,2,3] + [10]

↓

[11,12,13]
```

---

## Rule 3 --> Incompatible Shapes

If shapes are not compatible, NumPy raises an error.

Example

```python
(2,3)

+

(2,)
```

❌ ValueError

````python
# Rule 1

matrix = np.array([
    [1,2,3],
    [4,5,6]
])

vector = np.array([10,20,30])

print(matrix + vector)
````

## Explanation -->

Shape of matrix

```
(2,3)
```

Shape of vector

```
(3,)
```

NumPy automatically broadcasts the vector to each row.

Output

```python
[[11 22 33]
 [14 25 36]]
```

````python
# Rule 2

a = np.array([1,2,3])

b = np.array([10])

print(a + b)
````

## NOTE -->

`10` is automatically used for every element.

Output

```python
[11 12 13]
```

````python
# Rule 3

arr1 = np.array([
    [1,2,3],
    [4,5,6]
])

arr2 = np.array([1,2])

# print(arr1 + arr2)
# ERROR
````

## NOTE -->

Shape of `arr1`

```
(2,3)
```

Shape of `arr2`

```
(2,)
```

These shapes are **not compatible**.

Output

```python
ValueError:
operands could not be broadcast together
```

> **NOTE**
>
> `reshape()` may solve the problem **only if the new shape is compatible**. It is **not** a universal solution.

# 📘22.Vectorization



## NOTE -->

Vectorization means performing an operation on the **entire array at once** without using loops.

Advantages:

- No loops required.
- Less code.
- Faster execution than traditional loops.
- Used for mathematical and matrix operations.

````python
# Traditional Way (Using Loop)

list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = [x + y for x, y in zip(list1, list2)]

print(result)
````

## NOTE -->

This is the traditional Python way.

It works well for small data but becomes slower for large datasets.

````python
# Vectorization

import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([11, 23, 45])

result = arr1 + arr2

print(result)
````

## Explanation -->

The operation is applied to the **entire array at once**.

No loop is required.

````python
# Another Example

arr = np.array([10, 23, 45])

print(arr * 3)
````

## NOTE -->

Every element is multiplied by `3` automatically.

No loop is required.
