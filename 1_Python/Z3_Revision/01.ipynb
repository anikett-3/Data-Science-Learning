```python

import random

num = random.randint(10, 50)
i = 0

while i < 5:
    guess = int(input("Enter a random integer number between 10 to 50: "))

    if guess == num:
        print("You win, you guessed right!")
        break  # Game ends if correct guess
    elif guess > num:
        print("Your guess is too high, try a lower number.")
    elif guess < num:
        print("Your guess is too low, try a higher number.")

    i += 1  # Increment after each guess

if i == 5:
    print(f"You lose, you couldn't guess the correct number.\nThe correct number was: {num}")
```

```python
def check_if_prime(num):
    is_prime = True

    if num <=1:
        is_prime = False
    else:
        for i in range(2,int(num **0.5)+1):
            if num % i == 0:
                is_prime= False
                break
    if is_prime:
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not Prime Number")

check_if_prime(10)

```
