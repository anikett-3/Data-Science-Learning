# sets --   { }  curly braces

`SETS`  are built-in-data types in Python used to `store collection of unique items` . 
* They are unordered , meaning that the elements do not follow a specific order.
*  They do `not allow duplicate elements`.

* Sets are useful for membership tests, eliminating duplicate entries and-
* performing mathematical sets operations like `union, intersection, and symmetric difference`. 

```python
# create a set--->
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

* Mathematical operations -->

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## 1. UNION -->  combine given set.

union_set = set1.union(set2)

print(union_set)
```

```python
## 2. INTERSECTION ---. Give common elemeny from given element
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
## 3. INTERSECTION_UPDATE -->
# it perform operation and assign/update all new value in set1 

set1.intersection_update(set2)
print(set1)  # update value in set1.
print(set2)
```

```python
## 4. DIFFERENCE   -- subtract set1 from set2 OR remove common element from set1

set1 = {1,2,3,4,5,6}
set2= {4,5,6,7,8,9}

print(set1.difference(set2))  # subtract set1 from set2 OR remove common element from set1
print(set2.difference(set1))  # subtract set2 from set1 OR remove common element from set2
```

```python
## 5. DIFFERENCE_UPDATE -->
# it perform operation and assign/update all new value in set1 

set1.difference_update(set2)
print(set1)  # update value in set1.
```

```python
## 6. SYMMETRIC DIFFERENCE ..
# --> print unique / non common element from both given set 

print(set1.symmetric_difference(set2))
```

```python
set1 = {1,2,3}
set2 = {3,4,5}

## 1. IS subset     --> it checks element 

print (set1.issubset(set2))   # --> is set1 is subset of set2 -- false

print(set1.issuperset(set2))  # --> is set1 is super set of set2 -- false
```

```python
# if we want to remove duplicate element from list then we use "SET" -->

lst = [1,2,3,4,5,6,7,7,8,2,3,4,5]
set(lst)
```

```python
## 2. COUNTING UNIQUE word in text -->

text = "In this tutorial we are discussing about sets ."
words = text.split()      # split words 

# convert lists of words to set to get unique words

unique_word = set(words)
print(unique_word)
print(len(unique_word))
```
