```python
''' THis function is find num is odd or even '''

def even_or_odd(num):
    if num%2==0:
        print(f" {num} is Even number.")
    else:
        print(f" {num} is Odd number.")

# we reuse this code n mumber of time...

```

```python
# call this function --> 

even_or_odd(4)
```

```python
# FUNCTION with multiple parameters...

def add(a,b):   # function --1 
    c = a+b
    return c 

# OR --



def add_1(a,b):  # function --- 2
    return a+b 
```

```python
# calling above both function --

result = add(2,4) # function --1 
print(result)

# or 

print(add_1(3,6))  # function --- 2
```

> Positional And Keyword Arguments.

```python
''' if we have to give n number of parameters like this -->'''
#  def print_numebers(num1, num2 , num3, num4,)

''' IS it NOT  a lengthy SO 
 Instead of doing this we use 
 
 POSITIOAL ARGUMENTS ---> * args'''

def print_numbers(* args ): # inplace of " args " we write any thing like --> aman, qwert, asssdf, etc.
    for num in args:
        print(num)

# *args --> taking multiple parameters in one variable ...

# calling it 

print_numbers(1,2,3,4,5,6,"aman","kumar","etc", 3.14)     # only single *args are Refering all the multiple arguments given here.....


# This are called "POSITIONAL ARGUMENTS"


```

```python
''' KEYWORDS ARGUMENTS...  ---->'''
# it is given by   " **kwargs  "   .. double stars.  &   ... k = keywords ,, args = arguments 


# -- In keywords arguments all the Parameters will be in " KEY | VALUE " pairs 

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")


print_details(name= "aman", age = 22, Country = "India")

# kwargs take multiple parametrs in fomr of key value pairs..

```

```python
# NOW both in one block of code -->


def print_details_2(*args,**kwargs):

    for val in args:
        print(f" POSITIONAL ARGUMENTS -->> {val}")


    for key,value in kwargs.items():
        print(f" KEYWORDS ARGUMENTS -->> {key} : {value}")

```

```python
# giving argments in one line 

# and let decide function to SEPARATE THIS ---> in positonal & keywords arguments 

print_details_2(1,2,3,4,5,6,"amsem", "lodu", 3.14, name= "Abhishek kumar", age = "22", country = "India")

# look down it separate all the argument , measn it able to identify which arguments is from which parameters.
```
