```python
# WHILE loop

# while is continues to Excution as long as the conditon is true.

count = 0
while count<5:
    print(count)
    count=count+1
# count=count+1    # if count + 1 inn not under while then it is runnin in infinite loop.
```

```python
# LOOP control Statement.
#1. BREEAK --> Exit the loop prematurely.

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# 2. CONTINUE --> it skip the current the itration  and continue till last.

for i in range(10):
    if i%2==0:
        continue
        print(i)  # here print NOT work under continue.
    print(i)
```

```python
#3. PASS--->> pass statement is null operaton ;; it does NOTHING.

for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
## NESTED LOOPs

# a loops inside a loop.

for i in range(3):
    for j in range(2):
        print(f"i: {i} and j: {j}")
```

```python
## Examples-->

# print the first N Natural number using a while loops and for loops


## WHILE LOOPS

n = int(input("Enter a number: "))
count = 1
while count < n:
    print(count )
    count = count +1


## FOR LOOPS
for i in range (1,n+1):
    print(i)
```

```python
## 2. Examples-->

# calculate the sum of first N Natural number using a while loops and for loops


# WHILE LOOPS-
n = 10
sum = 0 
count = 1
while count <=n:
    sum = sum +count
    print(sum)       # this line print line by line execution.
    count = count +1

print(f"Sum of first N natural no is = {sum}")
```

```python
# FOR LOOPS-
n = 10
sum = 0
for i in range(1,n+1):
    sum = sum+i
print(f"Sum of first N natural no is = {sum}")
```

```python
## PRIME no --> between 1 to 100.

for num in range (1,101):
    if num > 1:
        for i in range (2, num):
            if num % i == 0:
                break
        else:
            print(num)
```
