```python
lst = [1,2,3,4]

print(lst)

lst[1] = "aniket"

print(lst)   ## SEE VALUE OF LIST AT INDEX 1 IS CHANGED .
```

* PACKING AND UNPACKING TUPLE

```python
# packing a tuple -----> " packed_tuple "
# it packed the value into  tuple -->

packed_tuple = 1,"hello",3.14
print(packed_tuple)
```

```python
# Unpacking tuple ----------> 
# it unpacked th tuple .

#  if we know already how many value have then we make exact no. of variable anf then unpacked ---:

a,b,c = packed_tuple

# here we know 3 value so 3 variable.
print(a)
print(b)
print(c)
```

```python
# UNPACKING WITH  * (STAR)

numb = (1,2,3,4,5,6)
first, *middle, last = numb    # *middle = "from 2nd to last-1 ""
print(first)
print(middle)
print(last)

```

```python
## NESTED TUPLE
# -->Tuples inside a tuples.

nstd_tpl = ((1,2,3,4),("A","b","v"),(1,"hello",3,12,"v"),(True,False))


# accessing the element inside the tuple-

print(nstd_tpl[0])    # accessing the 1st tuple
print(nstd_tpl[1][2])  # accessing the v from 2nd tuple
```

```python
# Iterating over the nested tuple

for sub_tuple  in nstd_tpl :
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
